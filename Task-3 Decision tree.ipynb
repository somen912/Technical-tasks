## Question- Create Decision Tree classifier on iris dataset and visualize it.

#### Import Library

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from matplotlib import pyplot as plt
from sklearn.metrics import accuracy_score
```

#### Load iris dataset

Importing Dataset from built in Iris dataset and creating Dataframe from it

```python
iris=load_iris()
dataframe=pd.DataFrame(iris.data,columns=iris.feature_names)
dataframe.head()
```

#### Prepare Data for the Decision Tree model

Preparing the training and testing data using the train_test_split function.

We are fitting the iris data and iris features to the train_test_split function.

```python
x=iris.data
y=iris.target
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
print("x train data", x_train.shape)
print("x test data", x_test.shape)
print("y train data", y_train.shape)
print("y test data", y_test.shape)
```

We are using 120 data for training the model and
30 data for testing the model.

#### Create Decision Tree and Train it

We are using the DecisionTreeclassifier and fit training features and training target to it.

```python
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
print("model is trained")
```

### Visualizing Decision Tree

For visualizing the Decision tree we are using plot_tree function.

```python
plt.figure(figsize=(20,10))
tree.plot_tree(model, filled=True)
```

# Testing The Model

To test our model we goona fit testing data features and the model gonna give us the predicted target values.

And we also see the comparison between the actual and the predicted value.

```python
prediction=model.predict(x_test)
prediction  # predicted value

df=pd.DataFrame({'predicted value':prediction,'actual value':y_test})
df['label']=df['predicted value'].replace(dict(enumerate(iris.target_names)))
df.head()
```

### Checking the accuracy

Now lets see how much accurate is our model.

```python
score=accuracy_score(prediction,y_test)
print(f"Accuracy score is {score} i.e. {score*100}%")
```
